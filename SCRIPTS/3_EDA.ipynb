```python
# Importing necessary packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading our csv file

df=pd.read_csv('DS4002-Project-3/DATA/Final_Dataset.csv')
```

```python
df.head()
```

```python
# Histogram of the indexes (which correspond to each of the 102 flower types)

plt.hist(df['Index'], bins=50,edgecolor='black')
plt.title('Histogram of Index')
plt.xlabel('Indexes (Corresponding to Each 102 Flower Types)')
plt.ylabel('Frequency')
plt.title('Histogram of Indexes (Corresponding to Each 102 Flower Types)')
```

```python
# Bar graph of all flower types

plt.figure(figsize=(20,6))
plt.hist(df['Name'], bins=100,edgecolor='black')
plt.title('Histogram of Flower Types')
plt.xlabel('Flower Types')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.title('Histogram of Flower Types')
```

```python
# Bar graph of most frequent flower types in dataset

flower_names = df['Name'].value_counts()
top_flowers = flower_names.nlargest(50)

plt.figure(figsize=(14,6))
plt.bar(top_flowers.index, top_flowers.values, edgecolor='black')
plt.title('Top 50 Most Frequent Flower Types')
plt.xlabel('Flower Types')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```
